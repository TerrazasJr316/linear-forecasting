## 01.- Extracción del DataSet

## 1.- Analísis Estadístico Descriptivo

```python
# Media en x

mean_x = x.mean()
mean_x
```

```python
# Media en y

mean_y = y.mean()
mean_y
```

```python
# Desviación Estandar en x
std_x = x.std()
std_x
```

```python
# Desviación Estandar en y
std_y = y.std()
std_y
```

```python
# Covarianza

covariance = df.cov().iloc[0, 1]
variance_x = x.var()
covariance
```

## 2.- Cálculo de Ecuación Normal

```python
# Ecuación Normal: Cálculo de b1 y b0

b1 = covariance / variance_x
b1
```

```python
b0 = mean_y - b1 * mean_x
b0
```

```python
y_pred = b0 + b1 * x
errors = y - y_pred
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x, y, label='Datos Observados')
ax.plot(x, y_pred, color='yellow', linewidth=2, label=f'Línea de Regresión (y = {b0:.2f} + {b1:.2f}x)')
ax.set_title('Regresión Lineal: Analisis Estadístico', fontsize=16)
ax.set_xlabel('Variable Independiente (x)', fontsize=12)
ax.set_ylabel('Variable Dependiente (y)', fontsize=12)
ax.legend()
ax.grid(False)
plt.show()
```

## 3.- Estimación de Errores

```python
# Estimación del error
y_pred = b0 + b1 * x
errors = y - y_pred
y_pred
```

```python
# Calculo y suma de errores medios
sse = (errors**2).sum()
sse
```

```python
sem = errors.sum()
sem
```

```python
# Eror estandar de estimación
eee = np.sqrt(sse / (n - 2))
eee
```

## 4.- Error Estándar de los Coeficientes b1 y b0

```python
ssx = ((x - mean_x)**2).sum()
eeb1 = eee / np.sqrt(ssx)
eeb1
```

```python
eeb0 = eee * np.sqrt((1/n) + (mean_x**2 / ssx))
eeb0
```

## 7.- Prueba de Hipotesis

```python
# Calculo de grados libertad
alpha = 0.05
print(f"Nivel de Confianza: {1-alpha:.0%}")
```

```python
grados_libertad = n - 2
grados_libertad
```

```python
# Región de rechazo
t_critico = stats.t.ppf(1 - alpha / 2, df=grados_libertad)
t_critico
```

### Estadístico de Contraste y Pruebas de Hipótesis en b1 y b0

Hay que tener en cuenta el siguiente analísis, para el calculo del *Estadístico de contraste* existiendo 2 casos:
* Si la hipótesis es nula en **b1** no hay una relación lineal sifniticativa ente x e y,
* Si la hipótesis de diferente de 0 en **b1** existe una relación lineal significativa.

```python
# Estadístico de Contraste y prueba de hipotesis en b1
t_b1 = b1 / eeb1
t_b1
```

```python
pH_b1 = 2 * (1 - stats.t.cdf(abs(t_b1), df=grados_libertad))
pH_b1
```

El siguiente cálculo se tomará en cuenta la ordenada origien **b0** al igual que en *b1* existen 2 casos:
* Si la hipótesis en **b0** es nula entonces la linea de regresión pasa por el origen.
* Si la hipótesis es diferente de 0 en **b1** significa que la ordenada es significativamente diferente de 0.

```python
t_b0 = b0 / eeb0
t_b0
```

```python
pH_b0 = 2 * (1 - stats.t.cdf(abs(t_b0), df=grados_libertad))
pH_b0
```

#### Conclusión:
*b1*

Al realiza la prueba en b1, se rechaza la hipótesis nula, ya que como se obvserva en el cálculo cumple con el nivel significativo de alpha que es el 5%

*b0*

Hecho el calculo se obeserva que la hipótesis no puede ser nula lo que quiere decir que es significativa estadísticamente.
